### Bivariate Exploration

#### What is the share of delayed flights? How does it split amongst different categories?

```python
# Create a dataframe that contains the frequency of delays per delay type and the respective quota
cols = ['ArrDelay','CarrierDelay','WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
df_delays = pd.DataFrame(index = ['Delay_Frequency','Delay_Quota'])
for col in cols:
    df_delays.loc['Delay_Frequency',col] = df[col][df[col] > 0].count()
    df_delays.loc['Delay_Quota',col] = df[col][df[col] > 0].count()/len(df)
df_delays.rename(columns={'ArrDelay' : 'OverallDelay'}, inplace=True)
```

```python
df_delays.iloc[0,:].plot(kind='bar', color = base_color);
# Add percentages
percentages = df_delays.loc['Delay_Quota', : ]
for i in range(len(percentages)):
    plt.text(x = i, y = df_delays.iloc[0,i]+100000, s = "{0:.2f}%".format(percentages[i]*100), horizontalalignment='center', color = "black")
plt.title('Delays per category')
plt.ylabel('Amount of flights')
plt.xlabel('Delay category')
plt.ylim(0,3500000);
```

The overall delay quota is 42.5 % which seems quite high. However any delay bigger than 0 minutes is considered here. The delay categories do not sum up to the total quota, as not for all delays the delay of these categories have been provided.

#### What is the relationship between departure delay and arrival delay?

```python
# In order to plot a scatter of the departure delay versus the arrival delay, first I only want to identify delays and not consider the early departures
df['ArrDelay2'] = df['ArrDelay'][df.ArrDelay > 0]
df.loc[df.ArrDelay <= 0, 'ArrDelay2'] = 0
df['DepDelay2'] = df['DepDelay'][df.DepDelay > 0]
df.loc[df.DepDelay <= 0, 'DepDelay2'] = 0
```

```python
# Plot a scatter plot of the departure delay versus the arrival delay
sns.set_style("darkgrid")
g = sns.regplot(data = df.sample(10000), x = 'DepDelay2', y = 'ArrDelay2', 
           fit_reg = False, scatter_kws = {'alpha' : 1/3})
g.figure.set_size_inches(10, 10)
X_plot = np.linspace(0, 1000, 100)
Y_plot = X_plot
plt.plot(X_plot, Y_plot, color='r');
```

As expected there seems to be a correlation between departure delay and arrival delay. However it can be seen that many points are above the red line (which has a slope equal to 1). That means that once the departure was delayed the arrival tends to be delayed even more. In my opinion this is unexpected as I assumed that delays can be caught up during the flight. A possible explanation for this is that an aircraft that is delayed in it's departure cannot get it's original timeslot in the arrival and therefore, gets even more delayed. One has to be careful with conclusions from this plot as the dots are highly overplotted. I will further investigate on this.

```python
# Introduce a new column that measures the difference between departure delay and arrival delay
df['DiffDepArr'] = df['ArrDelay2'] - df['DepDelay2']
```

```python
# Plot the histogram of the difference between arrival and departure delay
df.DiffDepArr[(df.DiffDepArr < 30) & (df.DiffDepArr > -30)].hist(bins=30);
```

```python
# Plot the histogram of the difference between arrival and departure delay without differences that equal to 0
df.DiffDepArr[(df.DiffDepArr < 30) & (df.DiffDepArr > -30) & (df.DiffDepArr != 0)].hist(bins=30);
```

In the first versus the second histogram it can be seen that the cases where the delay difference is equal to 0 are the most. From the second histogram without the 0 values it can be seen that the difference is quite evenly distributed towards right and left. This does not support the previous insight that departure delays cause significantly higher arrival delays.

#### Do different airports have different delay quotas considering the amount of flights they handled?

```python
# Calculate average delay time per departure airport versus amount of flights per origin airport
df2 = pd.DataFrame()
df2['OriginSumDepDelay'] = df.groupby(['Origin'])['DepDelay'].sum()
df2['OriginCountofFlights'] = df.groupby(['Origin'])['FlightNum'].count()
df2['OriginDelayperFlight'] = df2['OriginSumDepDelay'] / df2['OriginCountofFlights']
df2['OriginCountofDelFlights'] = df.Origin[df.DepDelay > 0].value_counts()
df2['ShareDelays'] = df2.OriginCountofDelFlights / df2.OriginCountofFlights
```

```python
g = sns.FacetGrid(data = df2, size = 5)
g.map(plt.scatter, 'ShareDelays', 'OriginCountofFlights', alpha = 1/2)
g.add_legend();
plt.ylabel('Count of flights per origin airport')
plt.xlabel('Delay quota (%)')
plt.title('Delay quota per origin airport');
```

As for the airports with a low number of flights the delay quota is spread out. For the airports with a higher amount of flights the delay quota is between 30% to 60%. The higher the flight frequency the less the spread of the delay quota.

#### What is the relationship between Actual elapsed time and arrival delay?

```python
# Plot the actual elapsed time versus the arrival delay
sns.regplot(data = df.sample(100000), x='ActualElapsedTime', y='ArrDelay2', scatter_kws = {'alpha' : 1/3}, line_kws = {'color' : 'red'});
```

Even though according to the dots it seems that lower elapsed time causes less delays in arrival the regression line suggests that there is a positive correlation between these two. This positive correlation is very low however. This leads to the conclusion that there is only a minor influence of the actual elapsed time on the arrival delay.

#### Are certain carriers sticking out regarding overall delays?

```python
# Calculate average delay time per carrier versus the amount of flights per carrier
df3 = pd.DataFrame()
df3['CarrierSumArrDelay'] = df.groupby(['UniqueCarrier'])['ArrDelay2'].sum()
df3['CarrierCountofFlights'] = df.groupby(['UniqueCarrier'])['FlightNum'].count()
df3['CarrierDelayperFlight'] = df3['CarrierSumArrDelay'] / df3['CarrierCountofFlights']
df3['CarrierDelayQuota'] = df.UniqueCarrier[df.ArrDelay > 0].value_counts() / df3.CarrierCountofFlights
```

```python
g = sns.FacetGrid(data = df3, size = 5)
g.map(plt.scatter, 'CarrierDelayperFlight', 'CarrierCountofFlights', alpha = 1/2)
g.add_legend();
plt.ylabel('Count of flights per carrier')
plt.xlabel('Average delay per flight (min)')
plt.title('Average delays per flight per carrier');
```

Apart from the two carriers with the lowest frequency on the left all the other carriers have an average delay ranging from 10 min per flight to 19 min per flight. There seems to be no relationship with regards to amount of flights per carrier and average delay per flight.

```python
g = sns.FacetGrid(data = df3, size = 5)
g.map(plt.scatter, 'CarrierDelayQuota', 'CarrierCountofFlights', alpha = 1/2)
g.add_legend();
```

The plot shows a similar result as the plot above, that the delay quota seems to have no relationship to the size of the carrier. The quota of delays is differing per carrier however as the values range from below 0.25 % up to almost 0.5 %.

#### Do delays bigger than 60 min occure more frequently during certain departure times or days of the week?

#### Are delays caused by weather more frequent and longer during certain months? How is the relationship between weather delays and overall arrival delays?

```python
plt.scatter(data = df.sample(100000), x = 'WeatherDelay', y = 'ArrDelay2', c = 'Month')
plt.colorbar();
```

From the visualization there can be no pattern identified with regards to weather delays in specific months. There are two lines formed by the points regarding the relationship between weather delays and overall arrival delays. One line is matching with the y axis. These are the overall delays which had no recordings or a value of 0 min for the weather delays. The other line has a slope of one. This is where the weather delay equals the overall arrival delay. As expected there is no point under this line as the arrival delay must be always at least as high as the weather delay.
